```python
import numpy as np
import matplotlib.pyplot as plt

def impermanent_loss(price_ratio):
    return 2 * np.sqrt(price_ratio) / (1 + price_ratio) - 1
```

```python
price_ratios = np.linspace(0.1, 5, 100)
il_values = impermanent_loss(price_ratios)

il_values[:5]
```

```python
plt.figure(figsize=(10, 5))
plt.plot(price_ratios, il_values)

plt.axhline(0)
plt.xlabel("Price change ratio")
plt.ylabel("Impermanent Loss")
plt.title("Impermanent Loss vs Price Change")

plt.show()
```

```python
def lp_fees_return(daily_volume_ratio, fee_rate, days):
    return daily_volume_ratio * fee_rate * days
```

```python
r = 2
il = impermanent_loss(r)

fees = lp_fees_return(
    daily_volume_ratio=0.10,
    fee_rate=0.003,
    days=30
)

il, fees
```
